```python
df.dropna(subset=['car name'], inplace=True)
display(df.isna().sum())
```

```python
df['mpg'] = df['mpg'].fillna(df['mpg'].median())
display(df.isna().sum())
```

```python
df['cylinders'] = df.cylinders.fillna(df.cylinders.mode()[0])
display(df.isna().sum())
```

```python
df['displacement'] = df.displacement.fillna(df.displacement.mean())
display(df.isna().sum())
```

```python
df['horsepower'] = df['horsepower'].interpolate()
df['weight'] = df['weight'].interpolate()
df['acceleration'] = df['acceleration'].interpolate()
display(df.isna().sum())
```

```python
df['model year'].fillna(method='bfill', inplace=True)
df['origin'].fillna(method='ffill', inplace=True)
display(df.isna().sum())
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

def qqp(data):
    data_sorted = np.sort(data)
    n = len(data_sorted)
    quantiles = (np.arange(1, n + 1) - 0.5) / n
    theoretical_quantiles = norm.ppf(quantiles)
    plt.figure(figsize=(8, 6))
    plt.scatter(theoretical_quantiles, data_sorted)
    plt.plot([min(theoretical_quantiles), max(theoretical_quantiles)], [min(data_sorted), max(data_sorted)], color='red', linestyle='--')
    plt.title('Q-Q Plot for MPG')
    plt.ylabel('Z values')
    plt.xlabel('Observed values')
    plt.grid(True)
    plt.show()
qqp(df['mpg'])
```

```python
import numpy as np
from scipy import stats
mean_mpg = df['mpg'].mean()
std_mpg = df['mpg'].std()
n = len(df['mpg'])
sem_mpg = std_mpg / np.sqrt(n)
confidence_level = 0.95
dof = n - 1
t_score = stats.t.ppf((1 + confidence_level) / 2, dof)
margin_of_error = t_score * sem_mpg
confidence_interval = (float(mean_mpg - margin_of_error), float(mean_mpg + margin_of_error))
print(f"The 95% confidence interval for the mean MPG is: {confidence_interval}")
```

```python
import numpy as np
from scipy import stats
mean_hp = df['horsepower'].mean()
std_hp = df['horsepower'].std()
n = len(df['horsepower'])
sem_hp = std_hp / np.sqrt(n)
cls = [0.90, 0.95, 0.99]
dof = n - 1
for cl in cls:
    ts = stats.t.ppf((1 + cl) / 2, dof)
    moe = ts * sem_hp
    print(f"The margin of error for the mean horsepower at {cl*100:.0f}% confidence level is: {moe:.4f}")
```

```python
print("Null Hypothesis (H0): There is no significant difference in the average MPG between U.S. and Japanese cars.")
print("Alternative Hypothesis (H1): There is a significant difference in the average MPG between U.S. and Japanese cars.")
```

```python
mpg_us = df[df['origin'] == 1]['mpg']
mpg_japan = df[df['origin'] == 3]['mpg']
display(mpg_us.head())
display(mpg_japan.head())
```

```python
from scipy.stats import ttest_ind
alpha = 0.05
t_statistic, p_value = ttest_ind(mpg_us, mpg_japan)
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
from scipy.stats import ttest_ind
t_statistic, p_value = ttest_ind(mpg_us, mpg_japan)
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
alpha = 0.05
print(f"Significance level (alpha): {alpha}")
print(f"Calculated p-value: {p_value:.4f}")

if p_value < alpha:
    print("Conclusion: Since the p-value is less than the significance level (alpha = 0.05), we reject the null hypothesis (H0).")
    print("There is a statistically significant difference in the average MPG between U.S. and Japanese cars.")
else:
    print("Conclusion: Since the p-value is not less than the significance level (alpha = 0.05), we fail to reject the null hypothesis (H0).")
    print("There is no statistically significant difference in the average MPG between U.S. and Japanese cars.")
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x='weight', y='mpg', data=df, scatter_kws={'alpha':0.5})
plt.title('Scatter Plot of MPG vs. Weight with Regression Line')
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
X = df[['weight']]
y = df['mpg']
model = LinearRegression()
model.fit(X, y)
```

```python
beta0 = model.intercept_
beta1 = model.coef_[0]
print(f"Regression Equation: MPG = {beta0:.4f} + {beta1:.4f}(Weight)")
```

```python
from sklearn.metrics import r2_score
y_pred = model.predict(X)
r2 = r2_score(y, y_pred)
print(f"R-squared score: {r2:.4f}")
```

```python
print(f"R-squared score: {r2:.4f}")
print("\nModel Limitations:")
print(f"- The R-squared score of {r2:.4f} indicates that only approximately 36.65% of the variance in MPG can be explained by the weight variable alone.")
print("- The scatter plot of MPG vs. Weight shows a non-linear relationship, suggesting that a simple linear model may not fully capture the underlying pattern.")
print("- Confounding factors not included in this model, such as engine size (displacement), horsepower, model year, and origin, likely influence MPG and could improve the model's predictive power if included.")
```
